```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_brasil_mortes, n_franca_mortes = 701494, 162906

data = {'Brasil':n_brasil_mortes, 'França':n_franca_mortes}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.bar(courses, values, color ='#fdfd96',
        width = 0.2)

plt.xlabel("Países")
plt.ylabel("Em milhões")
plt.title("Total de Mortes")
plt.show()
```

```python
n_brasil_casos = 37449418
n_franca_casos = 38930489

data = {'Brasil':37.44, 'França':38.93}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))


plt.bar(courses, values, color ='#5e2129',
        width = 0.2)

plt.xlabel("Países")
plt.ylabel("Em milhões")
plt.title("Total de Casos")
plt.show()
```

```python
labels = 'Brasil', 'França'
sizes = [n_brasil_casos, n_franca_casos]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Proporção de Casos por COVID-19 entre Brasil e França')
plt.show()
```

```python
import seaborn as sns
import pandas as pd


dados = {
    'País': ['Brasil', 'França'],
    'Mortes': [n_brasil_mortes, n_franca_mortes],
    'População': [210000000, 67000000],  # População aproximada em milhões
    'Mortes por Milhão': [n_brasil_mortes / 210, n_franca_mortes / 67]
}

df = pd.DataFrame(dados)

sns.scatterplot(data=df, x='População', y='Mortes', hue='País', size='Mortes por Milhão', sizes=(50, 200))
plt.title('Relação entre População, Mortes e Mortes por Milhão por COVID-19')
plt.xlabel('População (em milhões)')
plt.ylabel('Total de Mortes por COVID-19')
plt.show()
```
